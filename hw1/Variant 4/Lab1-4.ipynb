### 3. Визуализация

Используя библиотеки matplotlib/sns/plotly/pandas и набор данных представленный ранее визуализируйте

1. Как менялась средняя цена квартиры в Нагатинском Затоне(Nagatinski Zaton) с течением времени

````python

dataFrame["timestamp"] = pd.to_datetime(dataFrame["timestamp"])
nagatZaton = dataFrame[dataFrame["sub_area"] == "Nagatinskij Zaton"].copy()
nagatZaton["year"] = nagatZaton["timestamp"].dt.year
averagePrice = nagatZaton.groupby("year")["price_doc"].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")
sns.lineplot(data=averagePrice, x="year", y="price_doc", marker='o', color='darkblue', linewidth=2.5)
plt.title("Средняя цена квартиры в Нагатинском Затоне по годам", fontsize=16, fontweight='bold', color='darkgreen')
plt.xlabel("Год", fontsize=14, fontweight='bold', color='darkgreen')
plt.ylabel("Стоимость, млн руб", fontsize=14, fontweight='bold', color='darkgreen')
plt.xticks(rotation=45, ha='right', fontsize=12, color='darkgreen')
plt.xticks(ticks=averagePrice['year'], labels=averagePrice['year'].astype(int), fontsize=12, color='darkgreen')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
````

2. Постройте гистограмму распределения цен на квартиры с общей площадью более 50 кв.метров

````python
neededFlats = dataFrame[dataFrame["full_sq"] >50];
plt.figure(figsize =(10, 10))
sns.histplot(data=neededFlats["price_doc"], color="dodgerblue", kde=True, bins=50)
plt.title("Распределение цен квартир с площадью > 50 м²", fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel("Стоимость, млн * 10^2 руб", fontsize=14, fontweight='bold', color='darkblue')
plt.ylabel("Количество квартир", fontsize=14, fontweight='bold', color='darkblue')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show();
````

3. Визуализируйте количество проданных квартир в разных районах

````python
group = dataFrame["sub_area"].value_counts()
top_areas = group.head(15)
other_areas = group.tail(group.shape[0] - 10).sum()
top_areas["Прочие"] = other_areas
plt.figure(figsize=(8, 8))
plt.pie(top_areas, labels=top_areas.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors, 
        pctdistance=0.85, textprops={'fontsize': 10})
plt.title("Распределение квартир по районам (Топ-15)", fontsize=16, fontweight='bold', color='darkblue')
plt.show()
plt.show()
````

4. Постройте box plot отображающий распределение полных площадей квартир в Басманном районе

````python
data = dataFrame[dataFrame["sub_area"] == "Basmannoe"]["full_sq"]
print(data)
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, color="skyblue", fliersize=7)
plt.title("Распределение площади в Басманном районе", fontsize=16, fontweight='bold')
plt.ylabel("Площадь квартиры (м²)", fontsize=12)
plt.show()
````

5. Визуализируйте зависимость средней стоимости квартир в зависимости от этажа

````python
priceByFloor = dataFrame.groupby("floor")["price_doc"].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.lineplot(data = priceByFloor, x = "floor", y = "price_doc", marker='o', color='b')
plt.grid(True, linestyle='--', alpha=0.7)
plt.title("Средняя стоимость квартир в зависимости от этажа", fontsize=16, fontweight='bold')
plt.xlabel("Этаж", fontsize=12)
plt.ylabel("Средняя стоимость, млн руб", fontsize=12)
plt.show()
````

6. Покажите как меняется средняя площадь квартир в зависимости от района

````python
squareByArea = dataFrame.groupby("sub_area")["full_sq"].mean()
sns.set(style="whitegrid", palette="muted")

plt.figure(figsize=(24, 10))
sns.barplot(squareByArea)
plt.title("Средняя площадь квартир в зависимости от района", fontsize=18, fontweight='bold')
plt.xlabel("Район", fontsize=14)
plt.ylabel("Средняя площадь (м²)", fontsize=14)
plt.xticks(rotation=90)

for i, value in enumerate(squareByArea.values):
    plt.text(i, value + 0.5, f'{value:.1f}', ha='center', fontsize=12, color='black', rotation = 90)
    
plt.tight_layout()
plt.show()
````

7. Постройте график изменения медианной цены на квартиру менее 50 кв. метров за период с 2011 до 2014 года*

\* Задание на доп. балл

````python
dataFrame['timestamp'] = pd.to_datetime(dataFrame['timestamp'])
filtered_data = dataFrame[(dataFrame['full_sq'] < 50) & (dataFrame['timestamp'].dt.year >= 2011) & (dataFrame['timestamp'].dt.year <= 2014)].copy()
filtered_data['year_month'] = filtered_data['timestamp'].dt.to_period('M')
median_prices_monthly = filtered_data.groupby('year_month')['price_doc'].median()
plt.figure(figsize=(18, 6))
sns.lineplot(x=median_prices_monthly.index.astype(str), y=median_prices_monthly.values, marker='o', color='b', linewidth=2)
plt.title("Медианная цена квартир менее 50 кв. метров по месяцам (2011-2014)", fontsize=16, fontweight='bold')
plt.xlabel("Месяц", fontsize=14)
plt.ylabel("Медианная цена, млн руб", fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
````
